```python
%load_ext autoreload
%autoreload 1
%matplotlib inline
import matplotlib.gridspec as gridspec
import matplotlib.pylab as plt
import numpy as np
import h5py
import pickle
import time
from os import listdir
from os.path import isfile, join
import tensorflow as tf
from tqdm import tqdm_notebook
%aimport vae
```

## Main parameters

```python
npix = 21
batch_size = 120
n_latent = 2
nepochs = 10
```

## Using a small dataset for demonstration

```python
hdffile_path = "/home/phuijse/Work/Data/SUDARE_FAKE/dump/sudare_data_Feb8_npix21_r_SDSS_balanced.hdf5"

def print_attrs(name):
    print(name)
        
with h5py.File(hdffile_path, "r") as hf:        
    N_samples = hf['data/diff'].shape[0]
    hf.visit(print_attrs)
    big_batch_size = N_samples
    N_big_batches =  1#int(N_samples/big_batch_size)
    N_batches =   int(big_batch_size/batch_size)
    print("N_samples: %d Nbatches %d %d" %(N_samples, N_big_batches, N_batches))
    data = vae.normalize_batch(hf['data/diff'][:, :], normalize='median')
    labl = hf['data/labl'][:]
    RP = np.random.permutation(len(labl))
    train_data = data[RP[:120000], :]
    train_label = labl[RP[:120000]]
    test_data = data[RP[120000:], :]
    test_label = labl[RP[120000:]]
```

## Training the variational autoencoder

```python
# Train Importance Weighted autoencoder with 20 monte carlo realizations
model = vae.Autoencoder(variational=True, importance=True, mc_realizations=20,
                        npix=npix, latent_dim=n_latent, arch=2,
                        use_batch_norm=False, keep_prob=1.0, n_filters=128,
                        tag="nadam_decvar_relu", rseed=0)
print(model.get_number_of_parameters())
for k in tqdm_notebook(range(nepochs), total=nepochs, desc='AE train'):
    # Shuffle the big batch
    order = np.random.permutation(train_data.shape[0])
    N_batches = train_data.shape[0]//batch_size
    for j in range(N_batches):
        step = k*N_batches + j
        # Feed the mini batches
        mini_batch = train_data[order[j*batch_size:(j+1)*batch_size], :]
        cost  = model.partial_fit(mini_batch=mini_batch, step=step, train=True)

    N_batches = test_data.shape[0]//batch_size
    for j in range(N_batches):
        step = k*N_batches +  j
        mini_batch = test_data[j*batch_size:(j+1)*batch_size, :]
        cost  = model.partial_fit(mini_batch=mini_batch, step=step, train=False)

    if np.mod(k, 1) == 0:
        model.save_network(k)
```

## Generate latent space visualizations

```python
for ki, k in enumerate(range(0, nepochs)):    
    model_name = "models/IWAE_MC20_L2_npix21_ARCH2_Nf128_nadam_decvar_relu.ckpt-"+str(k)
    tf.reset_default_graph()
    saver = tf.train.import_meta_graph(model_name+".meta")

    sess = tf.InteractiveSession()
    saver.restore(sess, model_name)
    handles = sess.graph.get_collection('my_handles')
    input_data = handles[0]
    is_training = handles[1]
    encoder_mu = handles[4]
    decoder_mu = handles[5]
    encoder_lv = handles[6]
    z = handles[7]
    decoder_lv = handles[8]
    nz = 20
    d = 5
    z1_values = np.linspace(-d, d, nz)
    z2_values = np.linspace(-d, d, nz)

    canvas = np.empty((npix*nz, npix*nz))
    for i, z1 in enumerate(z1_values):
        for j, z2 in enumerate(z2_values):
            z_ = np.tile(np.array([[z2, z1]]), (1, 20, 1))
            x_mean = sess.run(decoder_mu, feed_dict={z: z_, is_training: False})
            canvas[(nz-i-1)*npix:(nz-i)*npix, j*npix:(j+1)*npix] = np.mean(x_mean[0, :, :], axis=0).reshape(npix, npix)

    fig = plt.figure(figsize=(15, 8), dpi=80)        
    ax = fig.add_subplot(1, 2, 1)

    ax.imshow(canvas, interpolation=None, cmap=plt.cm.RdBu_r, 
              extent=[-d, d, -d, d], vmin=-1., vmax=1.)


    code_mu, code_lv = sess.run([encoder_mu, encoder_lv], feed_dict={input_data: train_data, is_training: False})
    code_mu_t, code_lv_t = sess.run([encoder_mu, encoder_lv], feed_dict={input_data: test_data, is_training: False})
    ax = fig.add_subplot(1, 2, 2)

    from matplotlib.colors import LogNorm
    im = ax.hist2d(code_mu[:, 0], code_mu[:, 1], bins=40, norm=LogNorm(vmin=1e+0, vmax=1e+3), cmap=plt.cm.Reds)
    ax.set_xlim([-d, d])
    ax.set_ylim([-d, d])
    ax.set_aspect('equal')
    fig.subplots_adjust(right=0.85)
    #fig.subplots_adjust(bottom=0.15, left=0.05, right=0.85, top=0.7)
    cbar_ax = fig.add_axes([0.9, 0.15, 0.02, 0.7])
    fig.colorbar(im[3], cax=cbar_ax)
    
    plt.title("Iteration %02d" %(k))
    fig.savefig("thumbs/iwae_dvar%03d.png" %(ki))
    plt.close()
    sess.close()
```
